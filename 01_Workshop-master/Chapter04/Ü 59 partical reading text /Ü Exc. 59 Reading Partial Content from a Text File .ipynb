CONTROL STRUCTURE: E.G.  with() funtion does close a block of code automatically 

```python
#tells PY to read the first 30 caracters the whole tie
#with statement is a control flow structure 
with open('pg37431.txt') as f: 
    print(f.read(30))
```

```python
#reads line per line 
with open("pg37431.txt") as f:
    print(f.readline())
```
