```python
#import of libary 
import seaborn as sns
#list of data 
data = [90, 80, 50, 42, 89, 78, 34, 70, 67, 73, 74, 80, 60, 90, 90]
#used the density plot
sns.distplot(data)
```

```python
#chane titles 

import matplotlib.pyplot as plt
plt.title('Density Plot')
plt.xlabel('Score')
plt.ylabel('Density')
sns.distplot(data)
#important to not see the line of code 
plt.show()
```
