# Plotting a pie chart 

```python
#data input 
labels = ['Monica', 'Adrian', 'Jared']
num = [230, 100, 98]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
#pie chart have values and labels
plt.pie(num, labels=labels,autopct= "1.1f%%", colors= ['lightblue', 'lightgreen', 'lightyellow'])
plt.title("Voting Results: Club president", fontdict={"fontsize":20})
plt.show()
```
