```python
#Data 

grades = ['A', 'B', 'C', 'D', 'E', 'F']
students_count = [20, 30, 10, 5, 8, 2]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
plt.bar(grades, students_count,  color= ["green","lightgreen", "orange","lightpink","red", "purple"])
plt.title("count of studentts with specific grade", fontdict={"fontsize": 22})
plt.xlabel("grades")
plt.ylabel("count of students")
plt.show()
```

```python
plt.barh(grades, students_count,  color= ["green","green", "orange","orange","red", "red"])
plt.title("count of studentts with specific grade")
plt.xlabel("grades")
plt.ylabel("count of students")
plt.show()
```
