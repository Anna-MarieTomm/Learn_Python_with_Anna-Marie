```python
#find the correlation 

import seaborn as sns

sns.set()

corr = house_df.corr()
plt.figure(figsize=(8,6))
plt.title("correlation heatmap for Bosten housing data ", fontsize= 20)
sns.heatmap(corr, xticklabels = corr.columns.values, yticklabels = corr.columns.values, cmap = "Reds", linewidths = 1.5, alpha = 0.6 )

plt.show()
```
