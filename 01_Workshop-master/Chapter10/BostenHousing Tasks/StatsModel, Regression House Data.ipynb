```python
import pandas as pd 

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

house_df = pd.read_csv("HousingData.csv")
house_df
```

```python
#make a scatter plot with scatter()

x = house_df["RM"]
y = house_df ["MEDV"]


color = "green"
plt.figure(figsize=(15,10))
plt.xlabel("rooms", fontsize = 20)
plt.ylabel("median price", fontsize = 20)
plt.title("correlation: rooms and median price", fontsize= 20)
sns.regplot(x,y)

#in orange 
sns.residplot(x,y)
plt.scatter (x, y, c= color, alpha = 0.6)


plt.show()
```

```python
#get insides to the regession line 

#y = 0 

#confidence interval by default 95%

import statsmodels.api as sm 

#x = setting an intersept (with the smallest number of rooms)
X = sm.add_constant(x)
model = sm.OLS(y, X)
est = model.fit()
print(est.summary())
```
