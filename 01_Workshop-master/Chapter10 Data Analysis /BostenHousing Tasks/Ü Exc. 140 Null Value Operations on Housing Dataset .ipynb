# approches to deal with null values 

process known as: imputation of data 

- eleminate the rows (only if a few null values)
- replace with significant value, median or mean (only if the dataset is balanced)
- replace with the most likely value, for instance 0 or 1
- estimating the value with Mashine learning 


```python
#find the data with null in dataset

#counts how many 0 are in a respective column 
#any if one values or more are 0 
null_columns = housing_df.columns[housing_df.isnull().any()]

housing_df[null_columns].isnull().sum()
```

```python
#show a specific column and their respective nulls, shows the whole colums where the null occures 


housing_df[housing_df["CRIM"].isnull()][null_columns]


```

```python
#show all the columsn that are containing a 0 

housing_df[housing_df.isnull().any(axis=1)][null_columns].head()
```
