```python
import pandas as pd 

import matplotlib as plt

from matplotlib import pyplot as plot 

import seaborn as sns


#create a dataframe

smoking_df = pd.read_csv("china_smoking.csv")

smoking_df
```

```python
def my_histogram (column, title, xlab, ylab, bins = 10, alpha = 0.5, color = "c"): 
    plot.figure(figsize=(10,6))
    plot.hist(column, bins = bins, range = (80,800), alpha=alpha, color = color)
    plot.title (title, fontsize= 19)
    plot.xlabel(xlab)
    plot.ylabel(ylab)
    plot.savefig(title, dpi = 300)
    plot.show()

my_histogram(smoking_df ["smoking_yes_cancer_no"],"Smoking and chancer", "Chancer and Smoking", "Count" )
```
