```python
import pandas as pd 

housing = pd.read_csv("HousingData.csv")

housing.head()
```

```python
#find null values and replace them with median value


housing.loc [:,housing.isnull().any()]
```

```python
housing.describe()
```

```python
housing.median()
```

```python
housing = housing.fillna(housing.median())
housing
```

```python
#to check if therea are still values with 0 
housing.info()
```

```python

#import visualizatin platform
import matplotlib as plt

#make sure that you see it in juypter
%matplotlib inline


import seaborn as sns

#dark grid 
sns.set()

#plot histogram for the data 

#MEDV = median value of data set 

plt.pyplot.hist(housing ["MEDV"])

```

```python
plt.pyplot.title ("Median house prices in Bosten", fontsize=20)
plt.pyplot.hist(housing ["MEDV"])
plt.pyplot.xlabel("1980 Median Value in T")
plt.pyplot.ylabel("Count")
plt.pyplot.show()
plt.pyplot.figure(figsize=(10,6))

#saving the figure 
plt.pyplot.savefig("Median house prices in Bosten", dpi = 300)
plt.pyplot. show()
```
