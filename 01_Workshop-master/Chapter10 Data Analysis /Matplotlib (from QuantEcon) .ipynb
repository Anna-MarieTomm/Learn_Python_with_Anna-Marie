```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

```

MATLAB-style API (not prefered as it is rather un-Pythonic) 

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)

plt.plot(x, y, 'red', linewidth=4)
plt.show()
```

## Object-Oriented API (prefered option) 

- fig is a figure instance (black canvas) 
- ax is an Axes subplot instance (frame for plotting) 

```python


fig, ax = plt.subplots(figsize = (10,6), facecolor = "pink", edgecolor = "green")

#plot() option is a method of ax
ax.plot(x, y, 'r-', label = "sine function",  alpha = 0.4, linewidth=2)
ax.legend(loc = "lower right")
ax.set_yticks([-1,-0.5,0, 0.5,1])
ax.set_xticks([0,3,6,9,12])

ax.set_title ("Sinus curve", fontsize = 20)
plt.show()
```
