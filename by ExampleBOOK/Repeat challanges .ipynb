```python
#challange 120

import random 

def add(): 
    num1 = random.randint(5,20)
    num2 = random.randint(5,20)
    correctAnswer = num1 +num2 
    question1 = print ("\n" + str(num1)  + "+" + str(num2) + "= " )
    userAnwser = int(input("Input answer: "))
    answers = (correctAnswer, userAnwser)
    return answers

def subtract (): 
    num1 = random.randint(25,50)
    num2 = random.randint(1,25)
    correctAnswer = num1 - num2 
    question2 = print ("\n" + str(num1)  + "-" + str(num2) + "= " )
    userAnwser = int(input("Input answer: "))
    answers = (correctAnswer, userAnwser)
    return answers

def checker (userAnswer, correctAnswer):
    if userAnswer == correctAnswer: 
        print("Correct!")
        
    else: 
        print("Incorrect, the answer is: " + str(correctAnswer))
    
    
def main ():
    print("""
    1. Add
    2. Subtract
    
    """)
    choose = int(input("Enter 1 or 2: "))
    if choose ==1: 
        correctAnswer, userAnwser = add()
        checker(correctAnswer, userAnwser)
    elif choose == 2:
        correctAnswer, userAnwser = subtract()
        checker(correctAnswer, userAnwser)
    else: 
        print("your input is wrong! ")
    
main()
```

```python
#challange 140 
#work with subprograms

import sqlite3

#set up the database
with sqlite3.connect ("LiobaHouse.db") as db: 
    cursor= db.cursor()
    
def view():
    print ("---"*15)
    print ("\nAll Lioba Tenants: \n")
    cursor.execute("SELECT *FROM tenant")
    for row in cursor.fetchall(): 
        print(row )
    print ("---"*15)
    
def add (): 
    print ("---"*15)
    id_new = int(input("ID: "))
    name = input("Name: ")
    room = input("Room: ")
    city = input("Home country: ")
    cursor.execute("""INSERT INTO tenant (id, name, room, city)
                   VALUES (?,?,?,?)""", (id_new, name, room, city))
    db.commit()
    print ("---"*15)
    
def search(): 
    print ("---"*15) 
    which_room = input("W2hich room are you intersted in: ")
    cursor.execute ("""SELECT * FROM tenant WHERE room =?""", [which_room])
    for row in cursor.fetchall(): 
        print(row)
    print ("---"*15) 
    
def delete(): 
    print ("---"*15) 
    away = int(input("Which ID do you want to delete? "))
    cursor.execute ("DELETE FROM tenant WHERE id =?", [away])
    cursor.execute("SELECT * FROM tenant")
    db.commit()
    print ("---"*15) 
    
def main(): 
    print ("""\n
    1) View Lioba House
    2) Add to Lioba House
    3) Search for Tenant
    4) Delete Tenant 
    5) Quit\n""")

    go_on = True 
    while go_on == True:  
        selection = int(input("Enter your selection: "))
        if selection == 1: 
            view()
        elif selection ==2: 
            add()
        elif selection == 3: 
            search()
        elif selection == 4: 
            delete()
        elif selection == 5:
            print ("\nYou finished your SQL manipulation for today. Goodby!\n")
            go_on = False
        else: 
            selection = int(input("Wrong entry, try again: "))
    
main()
db.close()
```
