```python
#challange 128 (converter)

from tkinter import * 

#set up the window 
window = Tk()
window.geometry ("500x400")
window["bg"] = "yellow"
window.title ("distance converter")


#function
def convert_to_miles ():
    non_converted = entry.get()
    non_converted = int(non_converted)
    result = non_converted * 0.6214
    result = round (result, 2)
    converted ["text"] = (str(result) + "\nMiles") 

def convert_to_km (): 
    non_converted = entry.get()
    non_converted = int(non_converted)
    result = non_converted * 1.6093
    result = round (result, 2)
    converted ["text"] = (str(result) + "\nkm") 

#label (insert unit)
label = Label (text = "Enter unit")
label.place(x = 20, y = 50, width = 100, height = 30)
label ["bg"] = "green"
label ["fg"] = "white"

#entry unit 
entry = Entry (text = 0)
entry.place(x = 140, y = 50, width = 100, height = 30)
entry ["bg"] = "white"
entry ["fg"] = "blue"
entry.focus()



#button (from km to miles)
to_miles_button = Button (text = "to Miles", command = convert_to_miles)
to_miles_button.place(x = 260, y = 50, width = 100, height = 30)
to_miles_button ["bg"] = "purple"
to_miles_button ["fg"] = "blue"


#button (from miles to km)
to_km_button = Button (text = "to Kilometers", command = convert_to_km)
to_km_button.place(x = 260, y = 130, width = 100, height = 30)
to_km_button ["bg"] = "purple"
to_km_button ["fg"] = "blue"


#message (other unit)
converted = Message (text  = 0)
converted.place(x = 140, y = 130, width = 100, height = 60)
converted ["bg"] = "orange"
converted ["fg"] = "black"





window.mainloop()
```

```python
#challange 130 

from tkinter import * 
import csv 

#set up window 
window = Tk()
window.geometry ("500x400")
window["bg"] = "yellow"
window.title ("safe list to csv file")


#function (check if it is digit, than add if not clear entry box)
def diggit_add ():
    number = entry.get()
    number.isdigit() 
    if number.isdigit () == True: 
        number_List.insert (END, number)
        entry.delete (0,END)
        entry.focus()
    else:
        entry.delete(0,END)
        entry.focus()


def clear (): 
    number_List.delete(0,END)
    entry.focus()
    
def safe (): 
    file = open ("Numbers.csv", "w")
    temp_list = number_List.get(0,END)
    item = 0
    for row in temp_list: 
        newrecord = temp_list[item] + "\n"
        file.write (str (newrecord))
        item = item +1
        
        
    file.close()
    

#label (add number)
label = Label (text = "Enter a number:")
label.place (x = 20, y = 20, width = 100, height = 30)
label ["bg"] = "green"
label ["fg"] = "white"

#entry (number)
entry = Entry (text = 0)
entry.place(x = 140, y = 20, width = 100, height = 30)
entry ["bg"] = "white"
entry ["fg"] = "blue"
entry.focus()


#button (is digit, and add)
add_button = Button (text = "ADD, if digit", command = diggit_add)
add_button.place(x = 260, y = 20, width = 100, height = 30)
add_button ["bg"] = "purple"
add_button ["fg"] = "blue"

#button (clear)
clear_button = Button (text = "CLEAR", command = clear)
clear_button.place(x = 260, y = 70, width = 100, height = 30)
clear_button ["bg"] = "purple"
clear_button ["fg"] = "blue"

#button safe to .csv file
safe_button = Button (text = "SAFE, to csv", command = safe)
safe_button.place(x = 260, y = 120, width = 100, height = 30)
safe_button ["bg"] = "purple"
safe_button ["fg"] = "blue"



#list (with all digits)

number_List = Listbox ()
number_List.place (x = 140, y = 70,  width = 100, height = 80)

window.mainloop()
```
