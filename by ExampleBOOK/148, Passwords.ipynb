```python
import csv

#list with all data 

def create_file(): 
    file = list(csv.reader(open ("ID.csv")))
    temp_list = []
    for x in file:
        temp_list.append(x)
        print(x)

    return temp_list
    
    
def create_ID(temp_list): 
    name_again = True 
    while name_again == True: 
        your_ID = input("Which ID do you want to add? ")
        your_ID = str(your_ID)
        #check if ID in data already 
        
        inlist = False
        row = 0
        for y in temp_list: 
            if your_ID in temp_list[row] [0]: 
                print(your_ID, "has already been allocated")
                inlist = True
            row = row +1
        if inlist == False: 
            name_again = False
            
    return (your_ID)




def create_password():
    special_character = "!$§@%&<>*+#=?ß"
    try_again = True 
    while try_again == True: 
        score = 0
        password = input("\nGive me a passoword: ")
        lenght = len(password)
        #all criteria 
    
        criteria_one = False 
        criteria_two = False 
        criteria_three = False
        criteria_four = False
        criteria_five = False
    
        if lenght >= 8: 
             criteria_one = True 
        
        for x in password: 
            if x.islower(): 
                criteria_two = True 
         
            if x.isupper():
                criteria_three = True 
        
            if x.isdigit(): 
                criteria_four = True 
            
        
            if x in special_character: 
                criteria_five = True 
    
    
        if criteria_one == True:
            score = score +1
        if criteria_two == True: 
            score = score +1
        if criteria_three == True: 
            score = score +1
        if criteria_four == True: 
            score = score +1
        if criteria_five == True: 
            score = score +1
    
    
        print ("\nPassword score: ", score)
    
    
        if score <= 2: 
            print("Password is weak and will not be accepted... ")
            password = input("Give me a new one: ")
        
        elif score ==3 and score ==4: 
            print("Password is alright, but should be improved")
            improve = input("Do you want to improve the password? (yes/n0)) ")
            if improve == "yes" or improve == "Yes" or improve == "YES": 
                password = input("Your improved password, please: ")
            else: 
                print("Your choice, but the password is not optimal.")
                try_again = False
    
        elif score == 5:  
            print("\nYou created a very strong password!")
            try_again = False 
        return password
    
#option 2 change password
def search_user(temp_list): 
    ask_again = True
    your_ID = ""
    while ask_again == True: 
        search_ID = input("Which user ID do you want to change? ")
        in_list = False 
        row = 0 
        for y in temp_list: 
            if search_ID in temp_list [row][0]: 
                in_list = True
            row = row +1
            
        if in_list == True:
            your_ID = search_ID
            ask_again = False 
         
        else: 
            print(search_ID, "is not in list")    
    return your_ID


#first change in list than overwrite the document with the list! 
def change_password(your_ID, temp_list): 
    if your_ID != "": 
        password = create_password()
        ID = your_ID.index(your_ID)
        temp_list[ID] [1] = password 
        file = open ("ID.csv", "w")
        i =0
        for row in temp_list: 
            newRecord = temp_list [i][0]+ ", "+ temp_list [i][1]+"\n"
            file.write(str(newRecord))
            i = i+0
        file.close()



def display(): 
    print("\nAll IDs w/o passowords\n")
    
    temp_list = create_file()
    j = 0 
    for row in temp_list: 
        print(temp_list [j][0])
        j = j +1 


def main(): 
    temp_list = create_file()
    go_on = True
    
    while go_on == True: 
        
         #menue 
        print("""
        1. Create a new User ID
        2. Change a passoword
        3. Display all User IDs
        4. Quit Program 
    
        """)
    
        select = int (input("Enter a selection: "))
        if select == 1:
            your_ID = create_ID(temp_list)
            your_ID = str(your_ID)
            password = create_password()
            password = str(password)
            file = open("ID.csv", "a")
            newRecord = your_ID + ", " + password+ "\n"
            file.write(str(newRecord))
            file.close()
  
            
            
        elif select == 2: 
            your_ID = search_user(temp_list)
            change_password (your_ID,temp_list)
            
        elif select == 3:
            display ()
            
        elif select == 4: 
            print("\nYou are done with the project! Have a look at the CSV you have created")
            go_on = False
        else: 
            select = int(input("Your selection is out of the range, try it again: "))
            
            
main()
```
