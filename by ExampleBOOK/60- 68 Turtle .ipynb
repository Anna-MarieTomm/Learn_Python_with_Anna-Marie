```python
#challange 63 (3 squars with gap)

import turtle 

screen = turtle.Screen()
screen.bgcolor("red")

turtle.shape("turtle")
turtle.pensize(3)

for i in range (0,3): 
    filling = input("what color do you want: ")
    turtle.color("black", filling)
    turtle.begin_fill()
    turtle.pendown()
    for i in range(0,5):
        turtle.forward(90)
        turtle.right(90)
    turtle.penup()
    turtle.end_fill()
    turtle.left (90)
    turtle.forward (90)
       
    
turtle.exitonclick()
```

```python
#challange 66 (octagon = achteck )


#set up trutle
import turtle 
screen = turtle.Screen()
screen.bgcolor("yellow")

turtle.pensize (5)

#position setting
turtle.penup()
turtle.left(135)
turtle.forward(50)
turtle.pendown()

#set up random 
import random 


#paint octagon (1080)
angles = (8-2)*180
each_angle = angles/8
print ("the octagon has inner angles: ", angles, "and each inner-angle has: ", each_angle)

for i in range(0,8):
    colors = random.choice(["red", "blue", "black","white", "orange","purple"])
    turtle.color(colors,"")
    turtle.begin_fill()
    turtle.forward (100)
    turtle.right(45)
    
    turtle.end_fill()

#end program 
turtle.done()
```

```python
#challange 68 (beautiful flower )


#set up trutle
import turtle 
screen = turtle.Screen()
background = random.choice(["red", "blue", "black","white", "orange","purple"])
screen.bgcolor(background)

turtle.pensize (5)
import random

often = random.randrange(40,100)

for i in range (0,often): 
    color = random.choice(["red", "blue", "black","white", "orange","purple"])
    turtle.color(color,"") 
    angle = random.randrange(0,350,10)
    turtle.right (angle)
    long = random.randrange(0,250,20)
    turtle.forward(long)
    

turtle.done()
```
