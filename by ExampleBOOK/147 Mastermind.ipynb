```python
#first subprogram 

import random 


def computer_choice(): 
    print("\n\nYour options: yellow, green, red, blue\n")

    picker = []
    
    for i in range (0,4): 
        pick = random.choice(["yellow", "green", "red", "blue"])
        picker.append(pick)
    

    return (picker)

#select program 
def your(picker): 
    selection = []
    colorlist= ["yellow", "green", "red", "blue"]


    i=1
    while i <= 4: 
        select= input("Choose ONE color:\n " )
        if select in colorlist: 
            selection.append(select)
            i = i+1
        else: 
            select= input("try again, not in color list: \n")
            selection.append(select)
            i= i +1
        
    print("\nYour guess:\n", selection)
    
#checker program (right_right)

    right_right = 0 

    x=-1
    go_on = True
    while go_on == True: 
        x= x+1
        pick = picker[x]
        select = selection [x]
        if pick == select:
            right_right = right_right +1
        if x==3:
            go_on = False
    
    print ("\nCorrect colour in correct place: ", right_right)

    #checker right_wrong 
    right_wrong = 0 

    y = -1
    go_on_on = True 
    while go_on_on == True: 
        pick = picker[y]
        if pick in selection: 
            right_wrong = right_wrong + 1
            y = 1+y
        else: 
            y = 1+y
        if y == 3: 
            go_on_on = False

    print ("\nCorrect colour: ", right_wrong)

    almost = right_wrong - right_right

    print("\nCorrect colour BUT wrong place:", almost )
    data = (right_right,right_wrong, picker)
    return (data)


def main(): 
    trial = 0 
    picker= computer_choice()
    play = True
    while play == True:
        
        (right_right,right_wrong,picker) =your(picker)
    
        trial = trial +1
        if right_right == 4: 
            play = False
    print("\nYou win")
    print("\nIt took you:", trial, "trials")
    
        
main()
```
