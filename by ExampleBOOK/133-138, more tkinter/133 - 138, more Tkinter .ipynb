```python
#challange 135 

from tkinter import *

#set up window 
window = Tk()
window.geometry("500x300")
window.title("Color Game")


#functions, should change the background color of the window (without an if statement!)
def change_color(): 
    color = pickColor.get()
    window ["bg"] = color 
    window.update()

#drop-down list (with colors)
pickColor = StringVar (window)
pickColor.set ("Pick Color")
colorList = OptionMenu (window, pickColor, "blue", "red", "yellow", "white", "lightblue", "orange", "black", "lightgreen")
colorList. place (x= 20, y = 20)

#button (click me )

button = Button (text = "Click me!", command = change_color)
button.place(x = 230, y = 20, width = 100, height = 30)
button ["bg"] = "blue"

window.mainloop() 
```

```python
#challange 137

from tkinter import * 
window = Tk()
window.geometry("500x300")
window.title("Gender Joy")
window ["bg"] = "lightpink"


def add (): 
    name = name_entry.get ()
    gender = selectGender.get()
    newRecord = (name + ", " + gender + "\n")
    data_list.insert(END, newRecord)
    
    #all also to a text file
    file = open ("GenderName.txt", "a")
    file.write(newRecord)
    file.close()
    
    name_entry.delete(0,END)
    name_entry.focus()

    
def show_text_file(): 
    file = open ("GenderName.txt", "r")
    for row in file: 
        print (row)
    file.close()
    
    
#label (ask for name)
name_label = Label (text ="Input Name")
name_label.place(x = 20, y = 20, width = 100, height = 30)
name_label ["bg"] = "lavender"

#entry (name)
name_entry = Entry (text = "")
name_entry.place(x = 140, y = 20, width = 100, height = 30)
name_entry ["bg"] = "aquamarine"
name_entry.focus()

#dropdown list (gender)
selectGender = StringVar (window)
selectGender.set ("Select Gender")
genderList = OptionMenu (window, selectGender, "Female", "Male", "Choose not to say", "Transgender", "Not sure yet")
genderList.place(x = 140, y = 70)
genderList ["bg"] = "lightcoral"
genderList [ "fg"] = "darkolivegreen"

#listbox (name, gender, new line)
data_list = Listbox ()
data_list.place (x = 280, y = 20, width = 175, height = 100)
data_list ["bg"] = "goldenrod"

#button (ADD)
add_button = Button (text = "ADD to list", command = add)
add_button.place(x = 20, y = 70, width = 100, height = 30)
add_button ["bg"] = "skyblue"

#button (show textfile)
show_button = Button (text = "Show Text file", command = show_text_file)
show_button.place(x = 20, y = 120, width = 150, height = 30)
show_button ["bg"] = "wheat"


window.mainloop() 
```
