```python
#create the sql (both)

import sqlite3

#set up SQL

with sqlite3.connect("ArtGallery.db") as db: 
    cursor = db.cursor()
    
    
#create the ARTIST tabel 
    
cursor.execute ("""CREATE TABLE IF NOT EXISTS Artist (
ArtistID integer PRIMARY KEY,
Name text, 
Adress text , 
Town text,
Country text,
Postcode text);""")

cursor.execute(""" INSERT OR REPLACE INTO Artist (ArtistID, Name, Adress, Town, Country, Postcode)
VALUES ("1", "Martin Leighton", "5 Park Place", "Peterborough","Cambridgehire", "RE324 5LP")
""")
db.commit()



#create the PIECES TABLE 
cursor.execute( """ CREATE TABLE IF NOT EXISTS Pieces(
PieceID integer PRIMARY KEY, 
ArtistIDENTIFIER integer,
Title text, 
Medium text, 
Price integer);""")

cursor.execute("""INSERT OR REPLACE INTO Pieces (PieceID, ArtistIDENTIFIER, Title, Medium, Price)
VALUES ("1", "5", "Women with black Labrador", "Oil", "2200")
""")
db.commit()

cursor.execute("""INSERT OR REPLACE INTO Pieces (PieceID, ArtistIDENTIFIER, Title, Medium, Price)
VALUES ("7", "2", "Summer breeze", "Acrylic", "4000")
""")
db.commit()



#add artist 

add_artist_id = int(input("Artist ID: "))
add_artist_name = input("Artist Name: ")
add_adress = input("Artist Adress: ")
add_town = input("Arist Town: ")
add_country = input ("Country: ")
add_postcode = input("Artist Postcode: ")

cursor.execute (""" INSERT OR REPLACE INTO Artist (ArtistID, Name, Adress, Town, Country, Postcode)
VALUES (?,?,?,?,?,?)""",(add_artist_id,add_artist_name,add_adress,add_town,add_country,add_postcode))
db.commit()


#add piece

add_piece_id = int(input("Piece ID: "))
add_artist_identifier = input("Artist identifier: ")
add_titile = input("Piece title: ")
add_medium = input("Piece Medium: ")
add_price = int(input ("Price: "))

cursor.execute("""INSERT OR REPLACE INTO Pieces (PieceID, ArtistIDENTIFIER, Title, Medium, Price)
VALUES (?,?,?,?,?)""", (add_piece_id, add_artist_identifier, add_titile, add_medium, add_price))
db.commit()





#search (artist, piece and medium)
search_artist = input("Search: Artist name: ")
cursor.execute("""SELECT * FROM Artist WHERE Name =?""", [search_artist])
for x in cursor.fetchall(): 
    print(x)


search_piece = input("Search: Art piece: ")
cursor.execute("""SELECT * FROM Pieces WHERE Title =?""", [search_piece])
for x in cursor.fetchall(): 
    print(x)


search_medium = input("Search: Medium ")
cursor.execute("""SELECT * FROM Pieces WHERE Medium =?""", [search_medium])
for x in cursor.fetchall(): 
    print(x)



#buy function (removes it from SQL and safes it into a text file)
#add to text file 
buy = int(input("Select Piece ID of what you want to buy: "))
cursor.execute (""" SELECT * FROM Pieces WHERE PieceID = ?""", [buy])
for x in cursor.fetchall(): 
    file = open ("soldArt.txt", "a")
    newRecord = str(x) + "\n"
    file.write (str(newRecord))
    file.close()
    print(x)
    #add to thext file 
    
    #delete from sql 
cursor.execute("""DELETE FROM Pieces WHERE PieceID =?""", [buy])




#close the whole db
db.close()
```
