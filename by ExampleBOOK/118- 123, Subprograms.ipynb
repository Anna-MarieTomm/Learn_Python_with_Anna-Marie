```python
#challange 119 
import random 

def pick(): 
    low = int(input("pick a low: "))
    high = int(input("pick a high: "))
    comp_num = random.randint(low, high)
    return comp_num

def guess():
    guessi = int(input("I am thinking of a number, what do you think it is: "))
    return guessi 

def check(comp_num,guessi):
    correct = False
    while correct == False: 
        if comp_num == guessi: 
            print ("Well done! Your guess is mine")
            correct = True
        elif comp_num > guessi : 
            guessi = int(input("try again, your last number is too SMALL: "))
        else: 
            guessi = int(input("try again, your last number is too BIG: "))
        
def main (): 
    comp_num = pick()
    guessi = guess()
    check(comp_num,guessi) 
main()
```

```python
#challange 120

import random 


def addition (): 
    num_1 = random.randint(5,20)
    num_2 = random.randint(5,20)
    correct = num_1+ num_2
    print ("correct addition: ", correct)
    user_answer = int(input("Question: "+ str(num_1)+"+"+ str(num_2)+ "= "))
    #create tuple as only one variable can be transfered 
    answers = (correct, user_answer)
    return answers 

def subtraction (): 
    num_1 = random.randint(25,50)
    num_2 = random.randint(1,25)
    correct = num_1- num_2
    print ("correct substraction: ", correct)
    user_answer = int(input("Question: "+ str(num_1)+ "-"+ str(num_2)+ "= "))
    answers = (correct, user_answer)
    return answers 

def checker (correct, user_answer): 
    if correct == user_answer: 
        print ("well done")
    else: 
        print ("maybe next time... the real answer is: %d" %correct)

#main function
def main ():
    print ("\n1) Addition \n2) Subtraction\n------------------------------------------")
    pick = int (input ("Enter 1 or 2: "))
    print ("---"* 20)
    if pick ==1: 
        (correct, user_answer) = addition ()
        checker(correct, user_answer)
    elif pick ==2: 
        (correct, user_answer) = subtraction ()
        checker (correct, user_answer)
    else: 
        print("your choice is out of the range... ")
    
main()
```
