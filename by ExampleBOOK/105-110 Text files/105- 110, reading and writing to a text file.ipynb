```python
#example code

#write

file= open("Country.txt", "w")
file.write("Italy\n")
file.write("Germany\n")

file.close()


#read

file = open("Country.txt", "r")

print(file.read())


#append 

file = open("Country.txt", "a")
file.write("France\n")
file.close()

#read again 

file = open("Country.txt", "r")

print("new list: ")
print(file.read())
```

```python
#challange 109

print ("your options: \n 1) Create a new file\n 2) Display the file \n 3) Add a new item to the file ")

select = int(input("choose one of the options (1,2, or 3): "))


if select == 1: 
    #write file 
    subject = input("favorit school subject: ")
    file = open ("Subject.txt", "w")
    file.write(subject)
    file.write ("\n")
    file.close()

elif select == 2: 
    #display 
    file = open ("Subject.txt", "r")
    print(file.read())

    
elif select == 3: 
    add_subject = input("favorit other school subject: ")
    file = open ("Subject.txt", "a")
    file.write(add_subject)
    file.write ("\n")
    file.close()
    
    file = open ("Subject.txt", "r")
    print(file.read())
    
else: 
    print("please pic a valid number! ")
```
