```python
#challange 145 

from tkinter import *
import sqlite3

#set up the SQL 

with sqlite3.connect("TestScores.db") as db: 
    cursor = db.cursor()
cursor.execute ("""CREATE TABLE IF NOT EXISTS students(
name text PRIMARY KEY, 
score integer); """)
#commonads 

def add():
    name = entry_name.get()
    score = entry_score.get()
    cursor.execute ("""INSERT INTO students(name, score)
    VALUES (?,?)""", [name,score])
    db.commit()
    
def clear (): 
    entry_name.delete(0, END)
    entry_score.delete(0,END)
    entry_name.focus()

window = Tk()
window.geometry("350x250")
window ["bg"] = "purple"
window.title("TestScores")

label_name = Label (text = "Enter sudent´s name: ")
label_name.place(x = 20, y = 20, width = 150, height = 30)
label_name ["bg"] = "lightgreen"

label_score = Label (text = "Enter sudent´s score: ")
label_score.place(x = 20, y = 70, width = 150, height = 30)
label_score ["bg"] = "lightgreen"


entry_name = Entry (text = "")
entry_name.place(x = 190, y = 20, width = 100, height = 30)
entry_name.focus()

entry_score = Entry (text = "")
entry_score.place(x = 190, y = 70, width = 100, height = 30)

button_add = Button (text = "ADD", command = add)
button_add.place (x = 20, y = 120, width = 70, height = 30)

button_add = Button (text = "CLEAR", command = clear)
button_add.place (x = 110, y = 120, width = 70, height = 30)

window.mainloop()
```
